```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['employees.csv']))
df.head()
```

```python
#con esta funcion veremos los tipos que datos que se relacionan en la base de datos, aqui vemos que la mayoria de los datos son de tipo objeto, salario que es de tipo entero y bonus que es un flotante
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#con esta funcion de describe encontramos si tenemos problemas matematicos en la base de datos como extremos atipicos, que en este caso no los presentamos, ademas con describe nos muestra mas datos especificos
#como la media, el minimo, cuartiles, max y la desviacion estandar de la columna que en este caso es de las columnas que son numericas, ademas de count que nos muestra el numero de terminos en la columna.
df.describe()
```

```python
#como la funcion describe sin especificaciones solo mostrara columnas numericas agregando un "'include='all'" damos la indicacion de que nos muestre todos los datos segun los que arroja la funcion
#pero esta vez tomando encuenta los que no son numericos, mostrandonos todo el contenido de la base de datos pero con las condiciones facilitandonos tener datos mas especificos.
df.describe(include='all')
```

```python
#con la funcion info vamos a ver mas a detalle el contenido de la base de datos mostrandonos que columnas tiene contenida la base y los tipos de forma especifica.
df.info()
```

```python
#cambiamos todos los NaN a otro tipo de dato en este caso los cambiaremos a 0 y verificamos que si se hallan hecho los cambios de manera correcta, ahora se procede a hacer un analisis de si algunas
#columnas son necesarias para mantenerlas y si en caso tal de no serlo se procederian a borrarse, en este caso como tiene algunos datos importantes se mantendaran.
df.fillna(0, inplace=True)
df.head(20)
```

```python
#aqui como cambiamos todos los datos a 0 para cambiar en la columna genero de '0' a no definido hacemos lo siguiente para darle un especie de categoria 'no definido'
df['Gender'].replace(0,'no definido', inplace = True)
df['Team'].replace(0,'no definido', inplace = True)
```

```python
#aqui podemos ver la correlacion entre las columnas que son numericas, dando a entender el signo y la magnitud de la tendencia entre estas dos columnas.
df.corr()
```

```python
#para hacer este grafico debemos categorizar las variables dandoles un rango en que puedan estar asi haremos el grafico de barras mostrandonos donde se ubica cada dato
sns.stripplot(data=df, x="Gender", y="Salary")
```

```python
#como podemos ver en la grafica anterior no es comprensible, ya que tenemos una grafica con las columnas similares, por lo tanto para poder hacer mas entendible la grafica vamos a segmentar la columna 
#salario dandole un rango para poder hacer una grafica mas facil de entender.
bins = np.linspace(min(df["Salary"]),max(df["Salary"]),4)
grupo_nombre=["bajo","medio","alto"]
df["price_binned"]=pd.cut(df["Salary"], bins, labels=grupo_nombre,include_lowest=True)

df.head()
```

```python
#utilizando la libreria seaborn haremos uso de la funcion de grafico de barras para mostrar la informacion que cruzamos que en este caso lo hacemos a salario con genero dandonos el resultado.
sns.barplot(data=df, x="Gender", y="Salary")
#con este grafico podemos concluir que el salario no es algo que afecte el genero ni el genero afecta el salario ya que son muy similares los salarios, la unica diferencia es que las mujeres en promedio ganan
#un poco menos que los demas generos.
```

```python
#para ver la afectacion de los bonus que se dan por genero vemos que el genero no definido obtiene la mayor cantidad de bonus con sus salarios, en segundo lugar vemos a los hombre que obtienen un valor un poco
#menor, pero por ultimo vemos las mujeres que estan por debajo de los dos generos aunque sea un diferencia relativamente baja. Se puede concluir que el bonus si afecta al genero femenino.
sns.barplot(data=df, x="Gender", y="Bonus %")
```

```python
#ahora queremos hacer una compartiva de las areas que ganan mas dinero.
sns.barplot(data=df, x="Salary", y="Team")
#por medio de esta grafica podemos concluir una especie de escalafon, donde vemos que el area que mas dinero gana es el area de ingenieria y el que menos gana es servicio al cliente.
```

```python
#aqui si podemos ver que la afectacion del bonus es un poco marcada ya que si vemos en el grafico el area que mas bonus gana es el desarrollo de negocio, dejando en segundo puesto el que en el ejemplo anterior
#ganaba mas dinero que es ingenieria en tercer puesto y el que menos dinero gana ahora ocupa el segundo puesto que es servicio al cliente.
sns.barplot(data=df, x="Bonus %", y="Team")
```
